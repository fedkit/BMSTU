```python
import matplotlib.pyplot as plt

# Читаем данные из файла
processes = []
times = []

with open("answer.txt", "r") as f:
    for line in f:
        if line.strip():
            proc, time = line.strip().split("-")
            processes.append(int(proc.strip()))
            times.append(float(time.strip()))

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(processes, times, marker='o', linestyle='-', color='blue')
plt.xlabel("Количество процессов")
plt.ylabel("Время работы (сек)")
plt.title("Время работы MPI метода минимальных невязок от числа процессов")
plt.grid(True)
plt.xticks(processes)
plt.show()
```

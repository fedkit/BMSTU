```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads = []
times = []
```

```python
with open('answer2.txt', 'r') as file:
    for line in file:
        if '-' in line:
            parts = line.strip().split('-')
            threads.append(int(parts[0]))
            times.append(int(parts[1]))
```

```python
threads = np.array(threads)
times = np.array(times)
```

```python
plt.figure(figsize=(14, 6))
plt.plot(threads, times, 'bo-', linewidth=2, markersize=6)

plt.xlabel('Количество потоков', fontsize=12)
plt.ylabel('Время выполнения (мс)', fontsize=12)
plt.grid(True, alpha=0.5)

plt.yticks(range(min(times) - min(times) % 100, max(times) + 100, 500))

plt.xticks(threads)
plt.tight_layout()
plt.show()
```
